```python
#For each categorical column find unique values and number of times they appear:
#Set a function
def unique_values(col):
    u, c = np.unique(col, return_counts=True)
    for z in zip(list(u), list(c)):
        print('{}: {}'.format(z[0], z[1]))
```

```python
#Find the range of numerical columns
def range_values(col):
    c = min(col)
    d = max(col)
    return c, d
```

```python
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
import seaborn as sns
```

```python
#Pie chart for Boqcategory and RawConstructionData

labels = pd.value_counts(file_name['BOQ_Cat']).keys()
sizes = pd.value_counts(file_name['BOQ_Cat'])
properties = plt.pie(sizes, labels=labels, shadow=True, startangle=0, autopct='%1.1f%%')

for i in range(len(properties[0])):
    properties[1][i].set_size(13)
    properties[2][i].set_size(12)
plt.title("BOQ CATEGORY", size=25, fontweight='bold')
print
```

```python
#Length in RawConstructionData, we use dropna to dron the nan values.
#Assumption for the plot: The range in x-axis is [0,30], where there are very few around 40 and 50. 
#So, we keep the most important values for our distribution.
sns.distplot(file_name['Length'].dropna())
plt.title("LENGTH", fontweight='bold')
plt.xlim([0, 30])
print
```

```python
#Quantity-TotalCost
tmp = file_name[(file_name.Quantity < 10000) & (file_name.TotalCost < 50000)]
ax = sns.regplot(tmp.Quantity, tmp.TotalCost)
ax.set_facecolor((25/256, 0.9, 0.9))
```

```python
#Construction Element Types 

height = [3, 7, 2]
bars = ('Columns', 'ConcreteWall', 'Retaining Wall')
y_pos = np.arange(len(bars))
plt.bar(y_pos, height, color=(0.2, 0.4, 0.6, 0.6))
plt.xticks(y_pos, bars, color='black', rotation=70, fontsize='10', horizontalalignment='right')
 
plt.suptitle("Construction Element Types for Building Walls", fontweight='bold')
print
```

```python
#Boq category-total cost under 1000
tmp1 = file_name[(file_name.TotalCost < 1000)]
sns.barplot(x='BOQ_Cat', y='TotalCost', hue='BinQuantity', data=tmp1)
plt.xticks(rotation='vertical')
plt.title('Total Cost per BOQ Category', fontweight='bold')
```
